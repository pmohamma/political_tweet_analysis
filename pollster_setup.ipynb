```python
pprint(td[0]['data'])
```

```python
polls = [None]*len(td[0]['data'])
for i in range(len(td[0]['data'])):
    polls[i] = Poll(td[0]['data'][i])
    
candidate_set = set()
for p in polls:
    print(p.get_sorted_results())
    p.update_candidates(candidate_set)
```

```python
for i in candidate_set:
    print(i)
    for j in (i.polls):
        print(j)
    print("\n\n")
```
